```python
import pandas as pd
import string
import nltk
from nltk.corpus import stopwords
```

```python
tweets=pd.read_csv("tweets.csv",encoding="ISO-8859-1")
tweets.head()
```

```python
tweets['text'].head()
```

```python
punctuations=string.punctuation
def _clean(text):
    text=text.lower()
    text="".join(x for x in text if x not in punctuations)
    words=text.split()
    words=[w for w in words if w not in stopwords.words("english")]
    text=" ".join(words) 
    return text
```

```python
tweets['cleaned']=tweets['text'].apply(_clean)
```

```python
from collections import Counter
```

```python
complete_text=" ".join(tweets['cleaned'])
words=complete_text.split(" ")
Counter(words).most_common(50)
```

```python
raw_text=" ".join(tweets['text'])
words=raw_text.split()
mentions=[w for w in words if w.startswith("@")]
Counter(mentions).most_common(50)
```

```python
hashtags=[w for w in words if w.startswith("#") ]
Counter(hashtags).most_common(50)
```

```python
URLS=[w for w in words if w.startswith('http')]
Counter(URLS).most_common(50)
```

```python
from nltk import ngrams
bigrams=ngrams(complete_text.split(),2)
Counter(bigrams).most_common(50)
```

```python
from nltk import word_tokenize,pos_tag,ne_chunk
```

```python
for text in tweets['text']:
    entities=ne_chunk(pos_tag(word_tokenize(text)))
    for entity in entities:
        if hasattr(entity,"label"):
            print(entity)
```
